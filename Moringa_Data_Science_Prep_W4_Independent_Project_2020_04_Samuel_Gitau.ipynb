# IMPORTING LIBRARIES



```
# To import required libraries
# ---
# 
import pandas as pd
import numpy as np
```

# LOADING DATASETS


```
# Dataset 1
# ---
# Loading dataset 1 
# ---
# 
df = pd.read_csv('http://bit.ly/autolib_dataset')
df.head()
```

# DATA EXPLORATION


```
# To check information on our dataset
df.info()
```

```
# To understand our dataset
df.describe()
```

```
# To check for null values
df.isnull().sum()
```

# DATA CLEANING

```
#To remove white spaces in our columns
df.columns.str.replace(" "," ")
```

```
# To remove white spaces from our rows of string datatype


df['City']=df['City'].str.strip()
df['ID']= df['ID'].str.strip()
df['Kind']=df['Kind'].str.strip()
df['Public name']=df['Public name'].str.strip()
df['Rental status']=df['Rental status'].str.strip()
df['Status']=df['Status'].str.strip()
df.head(4)
```

```
# To drop 'Displayed comment' and 'Scheduled at' which have too many null values
new_df = df.drop(["Displayed comment","Scheduled at"],axis=1)
new_df
```

```
# To confirm columns with null values have been dropped
new_df.isnull().sum()
```

```
# To remove columns we don't require in analysis
new_df = df.drop(["Cars", "Address","Geo point","Charge Slots","Charging Status","Displayed comment", "Subscription status","Scheduled at", "Station type" ],axis=1)
new_df
```

```
# To remove duplicates from our dataset
df = new_df.drop_duplicates()
df
```

```
# To make our columns uniform

df.columns=df.columns.str.strip().str.upper().str.replace(" "," ")
df.head(1)
```

```
# To filter dataset according to station using ID, date using DAY column
# Reason-to get successive counters
df.sort_values(['ID', 'DAY'], ascending=[True, True], inplace=True)
df
```

# DATA ANALYSIS

```
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
df1 = df[['CITY', 'BLUECAR COUNTER', 'HOUR']]
df_paris = df[df['CITY'] == 'Paris']
df1 = df_paris.groupby(['BLUECAR COUNTER'])['HOUR'].max().sort_values(ascending=False)
df1
```

```
# What station is the most popular?
# Overall?
df_station = df.groupby(['ID'])['BLUECAR COUNTER'].max().sort_values(ascending=False)

df_station.head(1)
```

```
# What station is the most popular?
# At the most popular picking hour?
df_sta = df.groupby(['ID'])['HOUR'].max().sort_values(ascending=False)

df_sta.head(1)
```

```
# What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?
# Overall?
df_sta1 = df[['ID', 'POSTAL CODE', 'BLUECAR COUNTER']]
df_postal = df_sta1.sort_values('BLUECAR COUNTER', ascending=False)
df_postal.head(1)
```
